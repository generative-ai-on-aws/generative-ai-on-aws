### 제로샷 프롬프트

- 제로샷 프롬프트에서는 작업 예시 없이 단지 프롬프트 구조만을 모델에 제공합니다.

### 퓨샷 프롬프트
- 아래는 퓨샷(few-shot) 프롬프트의 예시입니다.
- 퓨샷 프롬프트에서는 프롬프트 구조뿐만 아니라 두 개 이상의 예시도 모델에 제공합니다.
- 프롬프트 구조와 예시를 활용하여 모델이 작업을 진행할 수 있는지 확인하기 위해 프롬프트를 제공하고 있습니다.


```python
prompt = """
Message: Hi Dad, you're 20 minutes late to my piano recital!
Sentiment: Negative

Message: Can't wait to order pizza for dinner tonight
Sentiment: Positive

Message: Hi Amit, thanks for the thoughtful birthday card!
Sentiment: ?
"""
response = llama(prompt)
print(response)
```

### 출력 형식 지정하기
- 모델이 응답할 형식을 지정할 수 있습니다,
- 아래 예시에서, "한 단어로 응답해 주세요"라고 요청할 수 있습니다.

```python
prompt = """
Message: Hi Dad, you're 20 minutes late to my piano recital!
Sentiment: Negative

Message: Can't wait to order pizza for dinner tonight
Sentiment: Positive

Message: Hi Amit, thanks for the thoughtful birthday card!
Sentiment: ?

Give a one word response.
"""
response = llama(prompt)
print(response)
```

- 더 큰 700억 개 파라미터 모델인 `llama-2-70b-chat`을 사용하여 보다 개선되고 확실한 응답을 얻을 수 있는지 확인해 보세요.

```python
prompt = """
Message: Hi Dad, you're 20 minutes late to my piano recital!
Sentiment: Negative

Message: Can't wait to order pizza for dinner tonight
Sentiment: Positive

Message: Hi Amit, thanks for the thoughtful birthday card!
Sentiment: ?

Give a one word response.
"""
response = llama(prompt,
                model="togethercomputer/llama-2-70b-chat")
print(response)
```

- 이제, 더 작은 모델을 다시 사용하되, 모델이 인스트럭션을 이해할 수 있도록 프롬프트를 조정해 보세요.
- 모델의 출력 형식을 `긍정적`, `부정적`, 또는 `중립` 중에서 선택하도록 제한해 보세요.

```python
prompt = """
Message: Hi Dad, you're 20 minutes late to my piano recital!
Sentiment: Negative

Message: Can't wait to order pizza for dinner tonight
Sentiment: Positive

Message: Hi Amit, thanks for the thoughtful birthday card!
Sentiment:

Respond with either positive, negative, or neutral.
"""
response = llama(prompt)
print(response)
```

### 연쇄적 사고 프롬프트
- 대규모 언어 모델은 문제를 더 작은 단계로 나누어 해결하도록 요청받으면, 더 나은 추론과 논리 문제 해결 능력을 발휘할 수 있습니다. 이를 **연쇄적 사고(chain-of-thought)** 프롬프트라고 합니다.

```python
prompt = """
15 of us want to go to a restaurant.
Two of them have cars
Each car can seat 5 people.
Two of us have motorcycles.
Each motorcycle can fit 2 people.

Can we all get to the restaurant by car or motorcycle?
"""
response = llama(prompt)
print(response)
```

- 모델에게 제공한 수학 문제를 "단계별로 생각해 보세요"라고 프롬프트를 수정합니다.






```python
prompt = """
15 of us want to go to a restaurant.
Two of them have cars
Each car can seat 5 people.
Two of us have motorcycles.
Each motorcycle can fit 2 people.

Can we all get to the restaurant by car or motorcycle?

Think step by step.
"""
response = llama(prompt)
print(response)
```

- 추가적인 인스트럭션과 함께 모델에게 제공합니다.

```python
prompt = """
15 of us want to go to a restaurant.
Two of them have cars
Each car can seat 5 people.
Two of us have motorcycles.
Each motorcycle can fit 2 people.

Can we all get to the restaurant by car or motorcycle?

Think step by step.
Explain each intermediate step.
Only when you are done with all your steps,
provide the answer based on your intermediate steps.
"""
response = llama(prompt)
print(response)
```

- 인스트럭션의 순서가 중요합니다!
- 모델에게 "먼저 답변하고" "나중에 설명해줘"라고 요청하여 출력 결과가 어떻게 달라지는지 확인해 보세요.

```python
prompt = """
15 of us want to go to a restaurant.
Two of them have cars
Each car can seat 5 people.
Two of us have motorcycles.
Each motorcycle can fit 2 people.

Can we all get to the restaurant by car or motorcycle?
Think step by step.
Provide the answer as a single yes/no answer first.
Then explain each intermediate step.
"""

response = llama(prompt)
print(response)
```

- 대규모 언어 모델은 답변을 한 번에 하나의 토큰씩 예측하므로, 단계별로 생각하도록 요청한 후에 답변을 제공받는 것이 최선의 방법입니다.
